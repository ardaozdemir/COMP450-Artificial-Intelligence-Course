```python
import numpy as np
import pandas as pd
from google.colab import drive 
from sklearn.feature_extraction import DictVectorizer


drive.mount('/content/gdrive')

df = pd.read_csv('gdrive/My Drive/dataset_turkey.csv' , sep = ",")

df = df.iloc[1:]

df = df[(df != 0).all(1)]

id_list = df["Code"].tolist()

for i in range(len(id_list)):

    id_list[i] = id_list[i][:2]

df["Code_ID"] = id_list

one_hot = pd.get_dummies(df["Code_ID"], prefix="code")
one_hot.head()

data = pd.concat([df, one_hot], axis=1)

data = data.drop(columns=['Code', 'Product label'])

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

y = df["Exported value in 2020"]

x = data.drop(columns=['Exported value in 2020'])


X1, X2, y1, y2 = train_test_split(x.values, y.values, random_state=0,
                                  train_size=0.75)
model = LinearRegression()
model.fit(X1, y1)

y_model = model.predict(X2)

score_test = model.score(X2, y2)

print(score_test)

from sklearn.model_selection import cross_val_score
cross_val_score(model, x, y, cv=5)
```

```python
from sklearn.ensemble import RandomForestRegressor

model_3 = RandomForestRegressor()
model_3.fit(X1, y1)
y_model_3 = model_3.predict(X2)

score_test_3 = model_3.score(X2, y2)

print(score_test_3)
cross_val_score(model_3, x, y, cv=5)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)

x_new = poly.fit_transform(x)

model_4 = LinearRegression().fit(x_new, y)
y_model_4 = model_4.predict(x_new)

score_test_4 = model_4.score(x_new, y)
print(score_test_4)
```
